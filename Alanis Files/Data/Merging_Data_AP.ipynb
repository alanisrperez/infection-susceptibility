```python
from google.colab import files
climate_file = files.upload()
```

```python
vaccine_df = pd.read_csv(io.BytesIO(vaccine_file['Vaccine_Data.csv']))
vaccine_df.head()
```

```python
vaccine_df['State'] = vaccine_df['State'].apply(lambda x: us.states.lookup(x).name if us.states.lookup(x) else x)
vaccine_df.head()
```

```python
vaccine_2020 = vaccine_df[vaccine_df['Year'] == 2020]
vaccine_2020.head()
```

```python
vaccine_2021 = vaccine_df[vaccine_df['Year'] == 2021]
vaccine_2021.head()
```

```python
vaccine_2022 = vaccine_df[vaccine_df['Year'] == 2022]
vaccine_2022.head()
```

```python
vaccine_2023 = vaccine_df[vaccine_df['Year'] == 2023]
vaccine_2023.head()
```

```python
death_df = pd.read_csv(io.BytesIO(death_file['Merged_Death_Census_Data.csv']))
death_df.head()
```

```python
clean_death_df = death_df.drop(columns=['Unnamed: 0', 'Covid Deaths Start Date', 'Covid Deaths End Date', 'Population 2023', 'Deaths from All Causes'])
clean_death_df.rename(columns={'County name': 'County'}, inplace=True)
clean_death_df.head()
```

```python
clean_death_df.rename(columns={'County name': 'County'}, inplace=True)
clean_death_df.head()
```

```python
census_df = pd.read_csv(io.BytesIO(census_file['census2020-2023.csv']))
census_df.head()
```

```python
census_df.rename(columns={'STNAME': 'State', 'CTYNAME': 'County'}, inplace=True)
census_df.head()
```

```python
census_2020 = census_df.drop(columns=['POPESTIMATE2021', 'POPESTIMATE2022', 'POPESTIMATE2023'])
census_2020.rename(columns={'POPESTIMATE2020': 'Population'}, inplace=True)
census_2020['Year'] = 2020
census_2020.head()
```

```python
census_2021 = census_df.drop(columns=['POPESTIMATE2020', 'POPESTIMATE2022', 'POPESTIMATE2023'])
census_2021.rename(columns={'POPESTIMATE2021': 'Population'}, inplace=True)
census_2021['Year'] = 2021
census_2021.head()
```

```python
census_2022 = census_df.drop(columns=['POPESTIMATE2020', 'POPESTIMATE2021', 'POPESTIMATE2023'])
census_2022.rename(columns={'POPESTIMATE2022': 'Population'}, inplace=True)
census_2022['Year'] = 2022
census_2022.head()
```

```python
census_2023 = census_df.drop(columns=['POPESTIMATE2020', 'POPESTIMATE2021', 'POPESTIMATE2022'])
census_2023.rename(columns={'POPESTIMATE2023': 'Population'}, inplace=True)
census_2023['Year'] = 2023
census_2023.head()
```

```python
climate_df = pd.read_csv(io.BytesIO(climate_file['counties20_23.csv']))
climate_df.rename(columns={'Name': 'County'}, inplace=True)
climate_df.head()
```

```python
climate_2020 = climate_df[['County', 'State', 'Avg 2020', 'Max 2020', 'Min 2020']]
climate_2020.rename(columns={'Avg 2020': 'Avg Temp', 'Max 2020': 'Max Temp', 'Min 2020': 'Min Temp'}, inplace=True)
climate_2020['Year'] = 2020
climate_2020.head()
```

```python
climate_2021 = climate_df[['County', 'State', 'Avg 2021', 'Max 2021', 'Min 2021']]
climate_2021.rename(columns={'Avg 2021': 'Avg Temp', 'Max 2021': 'Max Temp', 'Min 2021': 'Min Temp'}, inplace=True)
climate_2021['Year'] = 2021
climate_2021.head()
```

```python
climate_2022 = climate_df[['County', 'State', 'Avg 2022', 'Max 2022', 'Min 2022']]
climate_2022.rename(columns={'Avg 2022': 'Avg Temp', 'Max 2022': 'Max Temp', 'Min 2022': 'Min Temp'}, inplace=True)
climate_2022['Year'] = 2022
climate_2022.head()
```

```python
climate_2023 = climate_df[['County', 'State', 'Avg 2023', 'Max 2023', 'Min 2023']]
climate_2023.rename(columns={'Avg 2023': 'Avg Temp', 'Max 2023': 'Max Temp', 'Min 2023': 'Min Temp'}, inplace=True)
climate_2023['Year'] = 2023
climate_2023.head()
```

```python
merge_2020 = vaccine_2020.merge(census_2020, on=['County', 'State', 'Year'], how='left') \
               .merge(climate_2020, on=['County', 'State', 'Year'], how='left')
merge_2020.head()
```

```python
merge_2021 = vaccine_2021.merge(census_2021, on=['County', 'State', 'Year'], how='left') \
               .merge(climate_2021, on=['County', 'State', 'Year'], how='left')
merge_2021.head()
```

```python
merge_2022 = vaccine_2022.merge(census_2022, on=['County', 'State', 'Year'], how='left') \
               .merge(climate_2022, on=['County', 'State', 'Year'], how='left')
merge_2022.head()
```

```python
merge_2023 = vaccine_2023.merge(census_2023, on=['County', 'State', 'Year'], how='left') \
               .merge(climate_2023, on=['County', 'State', 'Year'], how='left')
merge_2023.head()
```

```python
merge_years = pd.concat([merge_2020, merge_2021, merge_2022, merge_2023])
merge_years.head()
```

```python
merge_final = merge_years.merge(clean_death_df, on=['County', 'State'], how='outer')
merge_final.tail()
```

```python
merge_final.to_csv('Merged_Data.csv', index=False)

# Download the CSV file
files.download('Merged_Data.csv')
```
