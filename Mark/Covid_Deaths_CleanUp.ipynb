```python
import pandas as pd
Covid_df = pd.read_csv('Covid_Death.csv')
Covid_df.head()
```

```python
Covid_df = Covid_df[['Start Date', 'End Date', 'State', 'County name', 'Deaths from All Causes']]
Covid_df.head(50)
```

```python
import us

# Apply the lookup function to the 'State' column to get the full state names
Covid_df['State'] = Covid_df['State'].apply(lambda x: us.states.lookup(x).name if us.states.lookup(x) else x)

# Display the result
Covid_df.head()
```

```python
Pop_df = pd.read_csv('co-est2023-alldata.csv', encoding='ISO-8859-1')
Pop_df.head()
```

```python
Pop_df = Pop_df[['STNAME','CTYNAME','POPESTIMATE2023']]
Pop_df.head(20)
```

```python
Org_Pop_df = []
for i, row in Pop_df.iterrows():
    if row['STNAME'] != row['CTYNAME']:
        Org_Pop_df.append(row)
```

```python
Org_Pop_df = pd.DataFrame(Org_Pop_df)
Org_Pop_df.head()
```

```python
Org_Pop_df['CTYNAME'].value_counts()
```

```python
Org_Pop_df.rename(columns={'STNAME': 'State','CTYNAME':'County name'}, inplace=True)
Org_Pop_df
```

```python
merged_df = pd.merge(Covid_df, Org_Pop_df, on=['County name', 'State'], how='inner')
merged_df.rename(columns={'Start Date':'Covid Deaths Start Date', 'End Date':'Covid Deaths End Date', 'POPESTIMATE2023':'Population 2023'}, inplace=True)
merged_df
```

```python
merged_df['Death Per 100k'] = merged_df['Deaths from All Causes'] / merged_df['Population 2023'] * 100000

merged_df
```
