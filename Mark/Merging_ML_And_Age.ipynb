```python
import pandas as pd
```

```python
ML_Data = pd.read_csv('ML_file.csv')
Age = pd.read_csv('cc-est2023-agesex-all.csv', encoding='ISO-8859-1')
```

```python
Age = Age.loc[Age['YEAR'] == 5]
Age.columns
```

```python
Age = Age[['STNAME','CTYNAME','POPESTIMATE','UNDER5_TOT','AGE6569_TOT','AGE7074_TOT', 'AGE7579_TOT', 'AGE8084_TOT', 'AGE85PLUS_TOT','MEDIAN_AGE_TOT']]
Age
```

```python
Age['65 plus per 100k'] = ((Age['AGE6569_TOT']+Age['AGE7074_TOT']+Age['AGE7579_TOT']+Age['AGE8084_TOT']+Age['AGE85PLUS_TOT']) / Age['POPESTIMATE']) * 100000
Age['5 Under per 100k'] = (Age['UNDER5_TOT'] / Age['POPESTIMATE']) * 100000
Age = Age[['STNAME','CTYNAME','65 plus per 100k','5 Under per 100k' ]]
Age = Age.rename(columns={'STNAME': 'State', 'CTYNAME': 'County'})
Age
```

```python
merged_df = pd.merge(ML_Data, Age, on=['State', 'County'], how='left')
merged_df
```

```python
merged_df.to_csv('ML_data_Age.csv', index=False)
```
