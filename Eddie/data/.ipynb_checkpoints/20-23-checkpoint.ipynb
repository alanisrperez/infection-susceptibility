```python
import pandas as pd

# Load each CSV file into a separate DataFrame
df_2020 = pd.read_csv('counties_2020.csv')
df_2021 = pd.read_csv('counties_2021.csv')
df_2022 = pd.read_csv('counties_2022.csv')
df_2023 = pd.read_csv('counties_2023.csv')

df_2020.head()
```

```python
df_2021.head()
```

```python
df_2022.head()
```

```python
df_2023.head()
```

```python
df_counties = pd.concat([df_2020, df_2021, df_2022, df_2023], axis=1)
df_counties.head()
```

```python
counties_df = df_counties.loc[:, ~df_counties.columns.duplicated()]
counties_df.head()
```

```python
counties_df.to_csv('counties20_23.csv', index=False)
```
