```python
import pandas as pd

# Load each CSV file into a separate DataFrame
df_avg = pd.read_csv('average_weather_data_2022.csv')
df_max = pd.read_csv('max_weather_data_2022.csv')
df_min = pd.read_csv('min_weather_data_2022.csv')

df_avg.head()
```

```python
df_avg = df_avg.drop(columns=['ID', 'Anomaly (1901-2000 base period)', 'Rank', '1901-2000 Mean', 'Unnamed: 7'], errors='ignore')
df_avg = df_avg.rename(columns={'Value': 'Avg 2022'})
df_avg.head()
```

```python
df_max.head()
```

```python
df_max = df_max.drop(columns=['ID', 'Anomaly (1901-2000 base period)', 'Rank', '1901-2000 Mean', 'Unnamed: 7'], errors='ignore')
df_max = df_max.rename(columns={'Value': 'Max 2022'})
df_max.head()
```

```python
df_min.head()
```

```python
df_min = df_min.drop(columns=['ID', 'Anomaly (1901-2000 base period)', 'Rank', '1901-2000 Mean', 'Unnamed: 7'], errors='ignore')
df_min = df_min.rename(columns={'Value': 'Min 2022'})
df_min.head()
```

```python
# Combine the DataFrames column-wise
combined_df = pd.concat([df_avg, df_max, df_min], axis=1)
combined_df.head()
```

```python
# Select only the first occurrence of "Name", "State", and the "Avg 2020", "Max 2020", and "Min 2020" columns
counties_2022 = combined_df.loc[:, ~combined_df.columns.duplicated()]
counties_2022.head()
```

```python
counties_2022.to_csv('counties_2022.csv', index=False)
```
